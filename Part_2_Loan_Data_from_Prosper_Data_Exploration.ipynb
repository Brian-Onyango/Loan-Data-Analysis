# Understanding the Profile Information of a Loan Borrower

## Investigation Overview

In this investigation, I wanted to understand the profile information of a loan borrower.

## Dataset Overview

This data set contains 113,937 observations with 81 variables of which the majority are numeric. The attributes used in this analysis are 'term', 'loanstatus', 'listingcategory (numeric)', 'borrowerstate', 'occupation', 'employmentstatus', 'employmentstatusduration', 'isborrowerhomeowner', 'statedmonthlyincome', 'loanoriginalamount', 'monthlyloanpayment', 'investmentfromfriendscount', 'investors'.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
from warnings import filterwarnings
filterwarnings('ignore')

%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('cleaneddataset.csv')
```

## Loan Status

Majority of the Loans are Current (>50000), followed by Completed (>30000) and Chargedoff (>10000) caps off top three.

```python
#loanstatus count plot
plt.suptitle("Title", y = 1.04)
plt.figure(figsize=[14, 6]);
sb.countplot(df['loanstatus'], palette = 'viridis', orient = 'v', order = df['loanstatus'].value_counts().index)
plt.title('Loan Status Count', fontsize=20, y = 1.05)
plt.xlabel('Loan Status', fontsize=15)
plt.ylabel('Count', fontsize=15);
```

## Borrowers State

CA (California) has the highest number of borrowers whereas ND (Nevada) has the least number of borrowers.

```python
# borrowerstate count plot
plt.suptitle("Title", y = 1.04)
plt.figure(figsize=[14, 6])
sb.countplot(df['borrowerstate'], palette = 'viridis', orient = 'v', order = df['borrowerstate'].value_counts().index)
plt.title('Borrower State Count', fontsize=20, y = 1.05)
plt.xlabel('Borrower State', fontsize=15)
plt.ylabel('Count', fontsize=15);
```

## Employment Status
Majority of the borrowers are Employed. Full-time employees makes the biggest percentage of the employed borrowes. At the bottom, the Retired tends to borrow less.


```python
# employmentstatus count plot
plt.suptitle("Title", y = 1.04)
plt.figure(figsize=[14, 6])
sb.countplot(df['employmentstatus'], palette = 'viridis', orient = 'v', order = df['employmentstatus'].value_counts().index)
plt.title('Employment Status Count', fontsize=20, y = 1.05)
plt.xlabel('Employment Status', fontsize=15)
plt.ylabel('Count', fontsize=15);
```

## Monthly Loan Payment vs Loan Original Amount

Monthly loan payment is positively correlated to loan original amount.

```python
def cor(col1,col2,d):
  ans = sb.regplot(x = col1, y = col2, data = d, scatter_kws = {"color": "red"}, line_kws = {"color": "black"})
  return ans
```

```python
# monthlyloanpayment and loanoriginalamount
plt.suptitle("Title", y = 1.04)
sb.set_style('whitegrid')
g =sb.lmplot(x='monthlyloanpayment',y='loanoriginalamount',data=df)
plt.title('Loan Original Amount vs Monthly Loan Paymnet', fontsize=20, y = 1.05)
plt.xlabel('Monthly Loan Paymnet', fontsize=15)
plt.ylabel('Loan Original Amount', fontsize=15)
g.fig.set_size_inches(14, 8);
```

## Monthly loan payment and Loan original amount grouped by loan Term

As much as the monthly loan paymnet is positively correlated to the loan original amount, it is interesting to note that there is a variance as per the steepness of the correlation. Loans with term limits of 12 tends to get lower loan original amount with a higher repayment amount compared to long term loans (60 months) which has a higher loan original amount with a lower monthly repayment amount.

```python
plt.suptitle("Title", y = 1.04)
g=sb.lmplot(x='monthlyloanpayment', y='loanoriginalamount', hue='term', 
           data=df.loc[df['term'].isin([12, 36, 60])], 
           fit_reg=False)
plt.title('Monthly Loan Payment vs Loan Original Amount grouped by Loan Term', fontsize=20, y = 1.05)
plt.xlabel('Monthly Loan Paymnet', fontsize=15)
plt.ylabel('Loan Original Amount', fontsize=15)
g.fig.set_size_inches(14, 8);
```

## Conclusion

>From this analysis, we can conclude that:

    * Majority of the borrowers are from CA (California)
    * Majority of the borrowers are Computer Programers
    * Majority of the borrowers are Employed
    * Majority of the borrowers are taking loans for Debt Consolidation
    * The median term for loans is 36 months
    * Majority of the borrowers are home owners
    * The average monthly income of the borrowers is 5694.34
    * The average amount of the loan given to the borrowers is 8587.75
    * The average monthly loan repayment is 279.58

>From the above insights, we can therefore profile an ideal candidate for a loan as:

* An employed Computer Programer from Californian who owns a home and has debts to consolidate and earns a monthly salary of $5694.34.
